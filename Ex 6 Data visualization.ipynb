# Aim:  To Perform Data Visualizatiom Using Matplotlib.

```python
#importing the basic library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

# Line chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.bar(x,y, color='purple')
plt.title("Bar Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

# Scatter Plot

```python
a=(1,5,8,6,3,7,9,4)
b=(10,55,21,64,85,33,44,28)

plt.scatter(a,b)
plt.title("Scatter Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
a=(1,5,8,6,3,7,9,4)
b=(10,55,21,64,85,33,44,28)

plt.scatter(a,b, color='purple')
plt.title("Scatter Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
a=(30,25,20,15,10)
plt.pie(a,labels=a)
plt.title("Pie Chart")
plt.show()
```

```python
a=(30,25,20,15,10)
b=('Apple','Banana','cherrie','dates','Elderberries')
plt.pie(a,labels=b)
plt.title("Pie Chart")
plt.show()
```
